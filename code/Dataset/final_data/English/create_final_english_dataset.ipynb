```python
import pandas as pd
import glob

# Step 1: Combine all 5 files into one
file_paths = ['audio_features_1-300.csv', 'audio_features_300-600.csv', 'audio_features_600-900.csv', 'audio_features_900-1500.csv', 'audio_features_1600-end.csv']  # Replace with your file names
combined_df = pd.concat([pd.read_csv(f) for f in file_paths], ignore_index=True)
combined_df = combined_df.drop_duplicates(subset='Song Title', keep='first')

combined_df.to_csv('combined_audio_features.csv', index=False)
```

```python
# For each Song tiltle in combined_df find the matching row in combined_file and append the lyrics column to combined_df

# Step 2: Read the combined file    
combined_file = pd.read_csv('combined_audio_features.csv')
# Step 3: Read the lyrics file
lyrics_file = pd.read_csv('combined_file.csv')

# Step 4: Merge the two dataframes on 'Song Title'
merged_df = pd.merge(combined_file, lyrics_file[['Song Title', 'English Lyrics']], on='Song Title', how='left')

merged_df = merged_df.drop_duplicates(subset='Song Title', keep='first')

merged_df.to_csv('merged_audio_features_lyrics.csv', index=False)
```
